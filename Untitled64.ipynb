```python
import numpy as np
import osqp
from scipy import sparse
import heapq
import matplotlib.pyplot as plt

# Функция для поиска пути с A*
def astar(start, goal, grid):
    def heuristic(a, b):
        # Евклидово расстояние
        return np.linalg.norm(np.array(a) - np.array(b))

    def get_neighbors(node):
        x, y = node
        neighbors = []
        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            nx, ny = x + dx, y + dy
            if 0 <= nx < grid.shape[0] and 0 <= ny < grid.shape[1] and grid[nx, ny] == 0:
                neighbors.append((nx, ny))
        return neighbors


    open_list = []
    heapq.heappush(open_list, (0 + heuristic(start, goal), 0, start))  # f, g, node


    g_score = {start: 0}
    came_from = {}

    while open_list:
        _, g, current = heapq.heappop(open_list)

        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1]

        for neighbor in get_neighbors(current):
            tentative_g_score = g + 1
            if neighbor not in g_score or tentative_g_score < g_score[neighbor]:
                g_score[neighbor] = tentative_g_score
                f_score = tentative_g_score + heuristic(neighbor, goal)
                heapq.heappush(open_list, (f_score, tentative_g_score, neighbor))
                came_from[neighbor] = current

    return []

# Размер и препятствия на сетке
grid = np.zeros((15, 15))
grid[5:10, 5:10] = 1  # Добавляем препятствия

# Начальная и конечная точки
start = (0, 0)
goal = (14, 14)

path_astar = astar(start, goal, grid)
path_astar = np.array(path_astar)
x_astar, y_astar = path_astar[:, 0], path_astar[:, 1]
n = len(x_astar)

# Матрица вторых разностей (для сглаживания кривизны)
D = sparse.diags([1, -2, 1], [0, 1, 2], shape=(n-2, n)).toarray()

# Q для минимизации кривизны
Q = sparse.block_diag([D.T @ D, D.T @ D], format='csc')

λ = 1.0  # Вес сглаживания
μ = 10.0  # Вес удержания маршрута

# Матрица A (ограничения: отклонение от исходного маршрута)
A = sparse.eye(2 * n, format='csc')

# Целевая функция (λ для сглаживания, μ для привязки к A*)
P = λ * Q + μ * A.T @ A
q = -μ * np.hstack([x_astar, y_astar])  # Сравниваем с исходным путем

l = np.hstack([-np.inf * np.ones(n), -np.inf * np.ones(n)])
u = np.hstack([np.inf * np.ones(n), np.inf * np.ones(n)])

prob = osqp.OSQP()
prob.setup(P, q, A, l, u, verbose=True)
res = prob.solve()

x_smooth, y_smooth = res.x[:n], res.x[n:]

plt.figure(figsize=(10, 10))
plt.imshow(grid.T, origin='lower', cmap='gray', extent=(0, grid.shape[0], 0, grid.shape[1]))
plt.plot(x_astar, y_astar, 'ro-', label="A* Path")
plt.plot(x_smooth, y_smooth, 'bo-', label="Smoothed Path")
plt.legend()
plt.show()
```
